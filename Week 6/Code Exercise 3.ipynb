    - Inside the function, calculate the commission using the following business rules (it is easy to implement the following using if, elif, else):
        - Up to 40 units sold = 5% commission

    - **Return instead of print the calculated commission** (Check the difference between value-returning function and void functions and make sure you return the value as a number.)

**The hard part of this problem is to know what should be done inside the function and what should be done outside the function. The comments in the code cell should help**

```python
# Enter your code here
def calculate_comission(num_of_units, unit_price=14):
    comission_rate = 0.
    if num_of_units <= 40:
        comission_rate = .05
    elif num_of_units <= 80:
        comission_rate = .06
    elif num_of_units <= 120:
        comission_rate = .07
    else:
        comission_rate = .08
    return unit_price * num_of_units * comission_rate


units_sold = int(input("Please enter units sold: ").strip())
if units_sold < 0:
    print("Please enter a positive integer")
else:
    comission_14 = calculate_comission(units_sold)
    print(
        f"For unit sold {units_sold} and unit price ${14} the comission is {comission_14}")
    comission_15 = calculate_comission(units_sold, 15)
    print(
        f"For unit sold {units_sold} and unit price ${15} the comission is {comission_15}")
```

## Problem 3 - Parse string for datetime
Write a program to read through the text file <b>"mbox-short.txt"</b>. Examine each line in the file in a loop. For the lines that start with "Date:", get the date and time part of the string, use datetime.strptime() to convert it to a datetime object and print out the result. Note that the datetime has 2 types of formats 
1. Sat, 05 Jan 2008 09:14:16
1. 2008-01-05 09:12:07

Your code needs to check which type the date is and use appropriate format string to perform the conversion. Be careful that space are not at the beginning or end of the string so the conversion can be successful. The needed module is imported for you in the cell already. DON'T REMOVE IT. Keep a counter to know how many dates are in this log. Print out the counter at the end.

```python
from datetime import datetime, date, timedelta
# Write your code below

with open("mbox-short.txt", 'r') as mbox:
    dates = 0
    dt_format_alpha = "%a, %d %b %Y %H:%M:%S"
    dt_format_digit = "%Y-%m-%d %H:%M:%S"

    for line in mbox:
        if not (line.startswith("Date:")):
            continue
        dates += 1
        date_line = line[line.index(' ') + 1: line.index("-0500")].strip()

        dt = None
        if date_line[0].isdigit():
            dt = datetime.strptime(date_line, dt_format_digit)
        else:
            dt = datetime.strptime(date_line, dt_format_alpha)
        print(dt)
    print(dates)
```
